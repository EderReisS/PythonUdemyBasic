```python
print("Por favor, insira sua altura e peso para prosseguir o c치lculo do IMC")
Altura = input("Digite sua altura em metros: ")
Peso = input("Digite seu peso em kg: ")
Altura = float(Altura)
Peso = float(Peso)
IMC = Peso/(Altura**2)
if IMC < 17 :
  print("Muito abaixo do peso ideal")
else:
  if IMC < 18.5:
    print("Abaixo do peso")
  else:
    if IMC < 25:
      print("Peso normal")
    else:
      if IMC < 30:
        print("Acima do Peso")
      else:
        if IMC < 35:
           print("Obesidade I")
        else:
          if IMC <40:
            print("Obesidade II (severa)")
          else:
            print("Obesidade III (m칩rbida)")                 
```
