Escreva um programa que receba dois números de ponto flutuante e mostre na tela o maior número digitado. Considere a possibilidade de o usuário digitar dois números iguais.

```python
print("Por favor escreva dois números: A e B(formato x.x)")
A = input("Digite o valor de A: ")
B = input("Digite o valor de B: ")

if A == B:
  print("Os valores de A e B são iguais!")
else:
  A = float(A)
  B = float(B)
  if A > B:
    print("Os valor de A é maior que B!")
  else:
    print("Os valor de B é maior que A!")
```
